```python
import pandas as pd
import sqlite3
import os

```

```python
os.listdir()

```

```python
users = pd.read_json("users.json")

print("Users loaded:", users.shape)
users.head()
```

```python
# Create SQLite database in memory
conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

# Execute SQL file
with open("restaurants.sql", "r") as file:
    cursor.executescript(file.read())

# Load restaurants table into DataFrame
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)

print("Restaurants loaded:", restaurants.shape)
restaurants.head()
```

```python
orders_users = pd.merge(
    orders,
    users,
    how="left",
    on="user_id"
)

print("After merging users:", orders_users.shape)
orders_users.head()
```

```python
final_df = pd.merge(
    orders_users,
    restaurants,
    how="left",
    on="restaurant_id"
)

print("Final dataset shape:", final_df.shape)
final_df.head()
```

```python
print("Orders rows:", orders.shape[0])
print("Final rows :", final_df.shape[0])
```

```python
final_df['order_date'] = pd.to_datetime(
    final_df['order_date'],
    format='%d-%m-%Y'
)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
print("final_food_delivery_dataset.csv created successfully")
```
